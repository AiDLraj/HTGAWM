```python
import os             
folder = "./Data/Antwon Michael Rose II/text/"
txt_files = os.listdir(folder)   # imagine you're one directory above test dir
print(txt_files)
```

```python
import pandas as pd
corpus = []

for file_path in txt_files:
    with open(folder+file_path,encoding='utf8') as f_input:
        corpus.append(f_input.read())

print(corpus)
```

```python
words_list = []

for phrase in corpus:
    words_list.append((str(phrase)).split())
print(words_list)
```

```python
words_corpus = []
for words in words_list:
    for word in words:
        words_corpus.append(word)
print(words_corpus)
```

```python
from collections import Counter

word_count = Counter(words_corpus)

print(word_count.most_common())
```
